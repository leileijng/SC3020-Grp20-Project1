```python
from datetime import datetime
import pandas as pd

date_format = "%d/%m/%Y"
start_date = datetime.strptime('1/1/2000', date_format)

number = df.shape[0]

concatFeature = [] 
# FG_PCT_home + TEAM_ID_home[-2:] + PTS_home + FT_PCT_home + AST_home + REB_home + HOME_TEAM_WINS

for i in range(number):
    now_date = datetime.strptime(df['GAME_DATE_EST'][i], date_format)  # Changed from [0] to [i]
    delta = now_date - start_date
    
    concatFeature.append(str(int(df['FG_PCT_home'][i] * 1000)).zfill(3) + 
                         str(df['TEAM_ID_home'][i])[-2:] + 
                         str(int(df['PTS_home'][i])).zfill(3) +
                         str(int(df['FT_PCT_home'][i] * 1000)).zfill(4) +
                         str(int(df['AST_home'][i])).zfill(2) + 
                         str(int(df['REB_home'][i])).zfill(2) +
                         str(df['HOME_TEAM_WINS'][i])
                        )

df['SEQ_NUMBER'] = concatFeature

# Drop rows where all elements are NaN
df.dropna(how='all', inplace=True)

# Save to a tab-separated text file
df.to_csv("games_final.txt", sep='\t', index=False)
```
